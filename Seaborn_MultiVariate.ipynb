```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('auto.csv')
auto = pd.DataFrame(data)
```

```python
sns.jointplot(x='hp', y='mpg', data=auto)
plt.title('Joint Distribution - Scatter')
plt.show()
```

```python
sns.jointplot(x='hp', y='mpg', data=auto, kind='hex')
plt.title('Joint Distribution-Scatter - Hexagonal Distribution')
plt.show()
```

```python
sns.jointplot(x='hp', y='mpg', data=auto, kind='kde')
plt.title('Joint Distribution-Scatter - Kernel Density Estimate')
plt.show()
```

```python
auto.head() #First five rows of the dataframe
```

```python
sns.pairplot(auto, vars=['mpg', 'hp', 'cyl'], hue='origin', kind='reg')
plt.show()
```

```python
corr_matrix = auto.corr()
print(corr_matrix)
```

```python
sns.heatmap(corr_matrix)
plt.title("Correlation Heat Map")
plt.show()
```

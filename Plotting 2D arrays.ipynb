```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generation_meshes():
    u = np.linspace(-2, 2, 41)
    v = np.linspace(-1, 1, 21)
    X, Y = np.meshgrid(u, v)
    Z = np.sin(3*np.sqrt(X**2 + Y**2))
    plt.pcolor(Z)
    plt.savefig('sine_mesh.png')
    plt.show()
```

```python
generation_meshes()
```

```python
def array_orientation():
    arr = ((np.array([[1, 2, 1], [0, 0, 1], [-1, 1, 1]])), 
       (np.array([[1, 0, -1], [2, 0, 1], [1, 1, 1]])), 
       (np.array([[-1, 0, 1], [1, 0, 2], [1, 1, 1]])), 
       (np.array([[1, 1, 1], [2, 0, 1], [1, 0, -1]])))
    for A in arr:
        plt.pcolor(A, cmap='Blues')
        plt.colorbar()
        plt.show()
```

```python
array_orientation()
```

```python
def contour_and_filled_contour():
    u = np.linspace(-2, 2, 101)
    v = np.linspace(0, 1, 51)
    X, Y = np.meshgrid(u, v)
    Z = np.sin(3*np.sqrt(X**2 + Y**2))
    
    plt.subplot(2,2,1)
    plt.contour(X, Y, Z)
    
    plt.subplot(2,2,2)
    plt.contour(X, Y, Z, 20)
    
    plt.subplot(2,2,3)
    plt.contourf(X, Y, Z)
    
    plt.subplot(2,2,4)
    plt.contourf(X, Y, Z, 20)
    
    plt.tight_layout()
    plt.show()
```

```python
contour_and_filled_contour()
```

```python
def modifying_colormaps():
    u = np.linspace(-2, 2, 101)
    v = np.linspace(0, 1, 51)
    X, Y = np.meshgrid(u, v)
    Z = np.sin(3*np.sqrt(X**2 + Y**2))
    
    plt.subplot(2,2,1)
    plt.contourf(X, Y, Z, 20, cmap='viridis')
    plt.colorbar()
    plt.title('Viridis')
    
    plt.subplot(2,2,2)
    plt.contourf(X, Y, Z, 20, cmap='gray')
    plt.colorbar()
    plt.title('Gray')
    
    plt.subplot(2,2,3)
    plt.contourf(X, Y, Z, 20, cmap='autumn')
    plt.colorbar()
    plt.title('Autumn')
    
    plt.subplot(2,2,4)
    plt.contour(X, Y, Z, 20, cmap='winter')
    plt.colorbar()
    plt.title('Winter')
    
    plt.tight_layout()
    plt.show()
```

```python
modifying_colormaps()
```

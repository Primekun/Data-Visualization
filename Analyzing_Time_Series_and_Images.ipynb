```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
aapl_data = pd.read_csv('apple_stocks.csv')
aapl_stock_date = aapl_data['date']
aapl_stock_close = aapl_data['close']
aapl_stock_date = pd.to_datetime(aapl_stock_date,infer_datetime_format=True)
aapl_frames = [aapl_stock_date, aapl_stock_close]
aapl = pd.concat(aapl_frames, axis=1)
aapl = aapl.set_index(['date'])
aapl = aapl.sort_index()

ibm_data = pd.read_csv('ibm_stocks.csv')
ibm_stock_date = ibm_data['date']
ibm_stock_close = ibm_data['close']
ibm_stock_date = pd.to_datetime(ibm_stock_date,infer_datetime_format=True)
ibm_frames = [ibm_stock_date, ibm_stock_close]
ibm = pd.concat(ibm_frames, axis=1)
ibm = ibm.set_index(['date'])
ibm = ibm.sort_index()

csco_data = pd.read_csv('cisco_stocks.csv')
csco_stock_date = csco_data['date']
csco_stock_close = csco_data['close']
csco_stock_date = pd.to_datetime(csco_stock_date,infer_datetime_format=True)
csco_frames = [csco_stock_date, csco_stock_close]
csco = pd.concat(csco_frames, axis=1)
csco = csco.set_index(['date'])
csco = csco.sort_index()

msft_data = pd.read_csv('msft_stocks.csv')
msft_stock_date = msft_data['date']
msft_stock_close = msft_data['close']
msft_stock_date = pd.to_datetime(msft_stock_date,infer_datetime_format=True)
msft_frames = [msft_stock_date, msft_stock_close]
msft = pd.concat(msft_frames, axis=1)
msft = msft.set_index(['date'])
msft = msft.sort_index()
```

```python
plt.plot(aapl, color='blue', label='AAPL')
plt.plot(ibm, color='green', label='IBM')
plt.plot(csco, color='red', label='CSCO')
plt.plot(msft, color='magenta', label='MSFT')

plt.legend(loc='upper left')
plt.xticks(rotation=60)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.xticks(rotation=45)
plt.title('AAPL: 2007 to 2017')
plt.plot(aapl, color='blue')

'''Slice aapl from '2012' to '2013' inclusive: view'''
view = aapl['2012':'2013']

plt.subplot(2,1,2)
plt.xticks(rotation=45)
plt.title('AAPL: 2012 to 2013')
plt.plot(view, color='black')

plt.tight_layout()
plt.show()
```

```python
view = aapl['2007-11': '2008-04']

plt.subplot(2,1,1)
plt.plot(view, color='red')
plt.title('AAPL: Nov. 2007 to Apr. 2008')
plt.xticks(rotation=45)

view = aapl['2008-01']

plt.subplot(2,1,2)
plt.plot(view, color='green')
plt.title('AAPL: Jan. 2008')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
view = aapl['2007-11': '2008-04']

plt.plot(aapl)
plt.xticks(rotation=45)
plt.title('AAPL: 2007-2017')

plt.axes([0.15, 0.5, 0.35, 0.35])
plt.plot(view, color='red')
plt.xticks(rotation=45)
plt.title('2007/11-2008/04')
plt.show()
```

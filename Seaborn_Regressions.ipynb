```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('auto.csv')
auto = pd.DataFrame(data)
```

```python
sns.lmplot(x='weight', y='hp', data=auto)
plt.title('Regression Between \nWeight and Horse Power)')
plt.show()

sns.residplot(x='weight', y='hp', data=auto, color='green')
plt.title('Residual Between \nWeight and Horse Power)')

plt.tight_layout()
plt.show()
```

```python
sns.lmplot(x='hp', y='mpg', data=auto)
plt.title('Regression Between \nHorse Power and MPG)')
plt.show()

sns.residplot(x='weight', y='hp', data=auto, color='green')
plt.title('Residual Between \nHorse Power and MPG)')

plt.tight_layout()
plt.show()
```

```python
plt.scatter(auto['weight'], auto['mpg'], label='data', color='red', marker='o')


sns.regplot(x='weight', y='mpg', data=auto, scatter=None, color='blue', label='order 1')

sns.regplot(x='weight', y='mpg', data=auto, scatter=None, order=2, color='green', label='order 2')

plt.legend(loc='upper right')
plt.show()
```

```python
sns.lmplot(x='weight', y='hp', hue='origin', palette='Set1', data=auto)
plt.show()
```

```python
sns.lmplot(x='weight', y='hp', hue='origin', row='origin', data=auto)
plt.show()
```
